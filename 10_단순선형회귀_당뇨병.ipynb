# 단순 선형회귀 - 당뇨병 사례

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- 당뇨병 데이터 셋 탐색

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
df.head()
```

```python
df.describe()
```

```python
print(diabetes.DESCR)
```

- 10개의 피쳐와 target간의 상관관계도(단순 선형회귀)

```python
import seaborn as sns
_, axs = plt.subplots(2, 5, figsize=(18,8))
for i, feature in enumerate(df.columns[:-1]):
    row, col = i // 5, i % 5
    sns.regplot(x=feature, y='target', data=df, ax=axs[row][col])
```

- feature vs target R squared 값

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
for feature in df.columns[:-1]:
    X = df[feature].values.reshape(-1,1)
    lr.fit(X, diabetes.target)
    score = lr.score(X, diabetes.target)
    print(f'{feature}: {score:.4f}')
```

### BMI vs target

```python
# Train/Test data set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    df.bmi.values.reshape(-1,1), diabetes.target, test_size=0.1, random_state=2022
)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# 가중치, 바이어스, 잔차 제곱의 합
lr.coef_, lr.intercept_, lr._residues
```

```python
# R squared
lr.score(X_test, y_test)
```

```python
# Mean Squared Error
from sklearn.metrics import r2_score, mean_squared_error
pred = lr.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = np.sqrt(mse)                     # Root Mean Squared Error
mse, rmse
```

- 시각화

```python
plt.scatter(X_train, y_train)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs target', fontsize=15)
plt.show()
```

```python
xs = np.array([-0.1, 0.18])
ys = xs * lr.coef_[0] + lr.intercept_
ys
```

```python
plt.scatter(X_train, y_train)
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs target', fontsize=15)
plt.show()
```

```python
# Test data 포함
plt.figure(figsize=(12,8))
plt.scatter(X_train, y_train, label='train dataset')
plt.scatter(X_test, y_test, marker='^', label='test dataset')
plt.plot(xs, ys, 'r-', lw=3)
plt.grid()
plt.legend()
plt.xlabel('BMI')
plt.ylabel('target')
plt.title('BMI vs target', fontsize=15)
plt.show()
```
